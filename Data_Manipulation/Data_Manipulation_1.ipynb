```python
import pandas as pd
import numpy as np
```

```python
low=pd.read_csv("/content/submission2.csv")
high=pd.read_csv("/content/submission1.csv")
```

```python
low.head()
```

```python
high.head()
```

```python
# Shuffle the DataFrame
shuffled_df = low.sample(frac=1, random_state=42).reset_index(drop=True)

# Calculate the split index
split_index = int(0.6 * len(shuffled_df))

# Split the DataFrame
df_60 = shuffled_df.iloc[:split_index]  # 60%
df_40 = shuffled_df.iloc[split_index:]   # 40%

# Display the results
print("60% Part:")
print(df_60)

print("\n40% Part:")
print(df_40)
```

```python
low_names = df_40['image_name'].tolist()
high_names = df_60['image_name'].tolist()
```

```python
print(high_names)
```

```python
new_df= low.copy()
new_df.head()
```

```python
total_changed_count = 0

for name in low_names:
    # Get the target value for the specified image_name from `low`
    target_value = high.loc[high['image_name'] == name, 'target'].values

    # Check if the target value exists and print it
    if target_value.size >= 0:
        # Update the target value in `new_df` DataFrame for the specified image_name
        new_df.loc[new_df['image_name'] == name, 'target'] = target_value[0]
        total_changed_count += 1
    else:
        print(f'{name} not found in the dataset low.')

print(total_changed_count, "values changed")
```

```python
new_df.head()
```

```python
new_df.columns = new_df.columns.str.strip()  # Removes any leading/trailing spaces
```

```python
are_equal = new_df.equals(low)
print(are_equal)
```

```python
are_equal = high.equals(low)
print(are_equal)
```

```python
new_df.to_csv('filename.csv', index=False)
```
